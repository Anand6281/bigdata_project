## Exploring the Dataset and Feature Engineering

Summary of the Data

```python
df = spark.read.csv("/content/GlobalLandTemperaturesByCity.csv", header=True, inferSchema=True)
```

```python
summary_df = df.describe()
summary_df.show()
```

List of Countries in Data

```python
unique_countries = df.select("Country").distinct().orderBy("Country")
unique_countries_list = [row.Country for row in unique_countries.collect()]
print(unique_countries_list)
```

Selecting only cities in United States

```python
df = df.filter(df["Country"] == "United States")
df = df.filter(df["City"] != "Anchorage")
```

Removing null value records

```python
df = df.na.drop(subset=["AverageTemperature"])
```

Adding new feature class_equator , timezone

```python
from pyspark.sql.functions import udf, when
from pyspark.sql.types import StringType

def extract_numeric(longitude):
    return float(longitude[:-1])

extract_numeric_udf = udf(extract_numeric, StringType())

df = df.withColumn("class_equator", extract_numeric_udf(df["Latitude"]))

df = df.withColumn(
    "class_equator",
    when((df["class_equator"] >= 0) & (df["class_equator"] < 30), "Class A")
    .when((df["class_equator"] >= 30) & (df["class_equator"] < 40), "Class B")
    .when((df["class_equator"] >= 40) & (df["class_equator"] <= 50), "Class C")
    .otherwise("Unknown")
)

df = df.filter(df["class_equator"] != "Unknown")
```

```python
def extract_numeric(longitude):
    return float(longitude[:-1])

extract_numeric_udf = udf(extract_numeric,StringType())
df = df.withColumn("timezone", extract_numeric_udf(df["Longitude"]))

df = df.withColumn(
    "timezone",
    when((df["timezone"] >= 112.5) & (df["timezone"] <= 127.5), "Pacific Time Zone")
    .when((df["timezone"] >= 97.5) & (df["timezone"] < 112.5), "Mountain Time Zone")
    .when((df["timezone"] >= 82.5) & (df["timezone"] < 97.5), "Central Time Zone")
    .when((df["timezone"] >= 67.5) & (df["timezone"] < 82.5), "Eastern Time Zone")
    .otherwise("Unknown")
)

df = df.filter(df["class_equator"] != "Unknown")
```

```python
from pyspark.sql.functions import col, when

def extract_month(dt):
    return int(dt[5:7])


extract_month_udf = udf(extract_month, StringType())
df = df.withColumn("month", extract_month_udf(df["dt"]))

df = df.withColumn(
    "season",
    when((df["month"] >= 3) & (df["month"] <= 5), "Spring")
    .when((df["month"] >= 6) & (df["month"] <= 8), "Summer")
    .when((df["month"] >= 9) & (df["month"] <= 11), "Fall")
    .otherwise("Winter")
)
```

List of Cites United States in Data

```python
unique_City = df.select("City").distinct().orderBy("City")
unique_City = [row.City for row in unique_City.collect()]
grouped_df = df.groupBy("City", "Latitude", "Longitude","class_equator","timezone").agg({"*": "count"}).toPandas()
print(unique_City)
```

**Finding Cities in world map using Latitude, Longitude**

```python
import folium

m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

for index, row in grouped_df.iterrows():
    city = row['City']
    latitude = float(row['Latitude'][:-1])
    longitude = float(row['Longitude'][:-1])
    lat_direction = row['Latitude'][-1]
    lon_direction = row['Longitude'][-1]

    if lat_direction == 'S':
        latitude = -latitude
    if lon_direction == 'W':
        longitude = -longitude

    folium.Marker([latitude, longitude], popup=city).add_to(m)

m.save('cities_map.html')
```

Finding Cities class_equator in world map using Latitude, Longitude

```python
import folium

m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)
class_colors = {
    "Class A": "blue",
    "Class B": "green",
    "Class C": "red"
}

for index, row in grouped_df.iterrows():
    city = row['City']
    latitude = float(row['Latitude'][:-1])
    longitude = float(row['Longitude'][:-1])
    lat_direction = row['Latitude'][-1]
    lon_direction = row['Longitude'][-1]

    if lat_direction == 'S':
        latitude = -latitude
    if lon_direction == 'W':
        longitude = -longitude

    class_equator = row['class_equator']
    color = class_colors.get(class_equator, 'gray')

    folium.Marker([latitude, longitude], popup=city, icon=folium.Icon(color=color)).add_to(m)


m.save('cities_equator_map.html')
```

Finding Cities Time Zone in world map using Latitude, Longitude

```python
import folium

m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

# Define colors for each time zone
timezone_colors = {
    "Pacific Time Zone": "blue",
    "Mountain Time Zone": "green",
    "Central Time Zone": "red",
    "Eastern Time Zone": "purple"
}

for index, row in grouped_df.iterrows():
    city = row['City']
    latitude = float(row['Latitude'][:-1])
    longitude = float(row['Longitude'][:-1])
    lat_direction = row['Latitude'][-1]
    lon_direction = row['Longitude'][-1]

    if lat_direction == 'S':
        latitude = -latitude
    if lon_direction == 'W':
        longitude = -longitude

    timezone = row['timezone']
    color = timezone_colors.get(timezone, 'gray')

    folium.Marker([latitude, longitude], popup=city, icon=folium.Icon(color=color)).add_to(m)

m.save('cities_Time_Zone_map.html')
```
